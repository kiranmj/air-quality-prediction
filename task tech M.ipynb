# As the dataset desciption states that the missing values are substituted by the value -200, we need to perform imputation to replace those values ,

```python
#in order to replace this NAN values with a a senseble value we do Central Imputation , 
##Hence, we groupby the date and take the mean and replace any NaN value by the mean of the values of that day.

df["T"] = df.groupby("Date")["T"].transform(lambda x: x.fillna(x.mean()))
```

```python

df["CO(GT)"] = df.groupby("Date")["CO(GT)"].transform(lambda x: x.fillna(x.mean()))
```

```python
df["NOx(GT)"] = df.groupby("Date")["NOx(GT)"].transform(lambda x: x.fillna(x.mean()))
df["NO2(GT)"] = df.groupby("Date")["NO2(GT)"].transform(lambda x: x.fillna(x.mean()))
df["RH"] = df.groupby("Date")["RH"].transform(lambda x: x.fillna(x.mean()))
```

```python
#here you still see the NULL because of their corresponding date vales are NULL values,
#so automatilly it took mean values also as a missing value.
#In order to overcome this issue we use funtion called called foreword fill i..e., "ffill" 

df.fillna(method='ffill', inplace= True)

```

#### We can also visualize the growth or decline of gas concentrations per month as plotted below

 As mentioned we have to check if the change in gas concentrations over time has any effect on the temperature change. So to do this, we need new variables which are the difference of consecutive values of gas concentrations compared with difference of values of temperature to compare if both differenced parameters have any relationship between them.

```python
df['dCO'] = df['CO(GT)'] - df['CO(GT)'].shift(-1)
```

```python
df['dT'] = df['T'] - df['T'].shift(-1)
```

```python
sns.stripplot(x="dT", y="dCO", data=df, jitter = True)
```

```python
sns.jointplot(x="dT", y="dCO", data=df, kind="reg")
```

### By the above graphs/plots we can say change in CO concentration effects the change in temperature
